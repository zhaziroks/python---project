```python
weather = pd.read_excel('weather.xls')
weather.head()
weather = weather.iloc[1:]
```

```python
weather = weather.iloc[1:].reset_index(drop=True)
weather.head()
```

```python
weather['Дата'] = pd.to_datetime(
    weather['Дата'],
    format='%d.%m.%Y %H:%M'
)
```

```python
weather['Дата'] = weather['Дата'].dt.date
```

```python
weather['T'] = pd.to_numeric(weather['T'], errors='coerce')
```

```python
weather_daily = weather.groupby('Дата')['T'].mean().reset_index()
```

```python
weather_daily.head()
```

```python
grouped_df['Дата'] = pd.to_datetime(grouped_df['Дата']).dt.date

merged = pd.merge(grouped_df, weather_daily, on='Дата')

merged.head()
```

```python
fig, ax1 = plt.subplots()

ax1.plot(merged['Дата'], merged['Количество'], color='blue')
ax1.set_ylabel('Количество продаж')
ax1.set_xlabel('Дата')

ax2 = ax1.twinx()
ax2.plot(merged['Дата'], merged['T'], color='red')
ax2.set_ylabel('Температура')

plt.title('Зависимость продаж и температуры')
plt.show()
```

```python
#На графике видно, что продажи и температура изменяются во времени. 
#Явной линейной зависимости между температурой и количеством продаж не наблюдается.
```
